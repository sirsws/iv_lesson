```python
#待用函数知识简介
##改写一下上面程序
#information_list=[['第一条','今天上课'],['第二条','冬奥会开幕'],['第三条','冬残奥会开幕']]
information_list=['今天上课','冬奥会开幕','冬残奥会开幕']

##这里len是取长度，range是取范围

# print(len(information_list))

# b=0
# c=0
# for a in range(len(information_list)):
#     print('我是range值'+str(a))
#     if '冬奥' in information_list[a]:
#         print(str(a)+'冬奥相关')
#         b=b+1
#     elif '冬残奥' in information_list[a]:
#         print(str(a)+'冬残奥相关')
#         c=c+1
#     else:
#         print(str(a)+'不知道是啥')
        
#程序函数化

def w_olympic(information_list):
    b=0
    c=0
    for a in range(len(information_list)):
        #print('我是range值'+str(a))
        if '冬奥' in information_list[a]:
            print(str(a)+'冬奥相关')
            b=b+1
        elif '冬残奥' in information_list[a]:
            print(str(a)+'冬残奥相关')
            c=c+1
        else:
            print(str(a)+'不知道是啥')
    return b,c




information_list=['今天上课','冬奥会开幕','冬残奥会开幕','冬奥会冰墩墩热卖','中兵红箭','乌克兰','冬残奥卖雪融容']
d=w_olympic(information_list)

print(str(d[0])+'条冬奥相关')
print(str(d[1])+'条冬残奥相关')
```

```python
#requests库和re库简单示例程序
#功能：一个简单的新闻检索程序，监测百度关键词搜索新闻内容中冬奥和冬残奥相关新闻数量
#可以通过更换keyword变量内容调整关键词
#冬奥和冬残奥替换功能没有做，入需替换成其他可以全文替换对应内容

##这段后面讲
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Safari/537.36'}
keyword='雪融容'
url = 'https://www.baidu.com/s?tn=news&rtt=4&bsst=1&cl=2&wd=' + keyword
res = requests.get(url, headers=headers, timeout=10,verify=False).text
p_total='<div><h3 class="news-title_1YtI1"><a href="(.*?)" target="_blank" class="news-title-font_1xS-F" aria-label="标题：(.*?)" data-click="{.*?}">.*?aria-label="新闻来源：(.*?)">.*?</span>\s*<span class="c-color-gray2 c-font-normal" aria-label="发布于：(.*?)">.*?</span>\s*</div><span class="c-font-normal c-color-text" aria-label="摘要\s*(.*?)">'
total=re.findall(p_total,res,re.S)

total
total[0][4]
##建立一个空的list存放雪融容新闻信息
xrr_informations=[]
for i in range(len(total)):
##用append将信息内容加在原list最后
    xrr_informations.append(total[i][4])
print(xrr_informations)

#将雪融容新闻信息列表代入上面函数
d=w_olympic(xrr_informations)

print(str(d[0])+'条冬奥相关')
print(str(d[1])+'条冬残奥相关')
    


```
