```python
# encoding:utf-8
import requests 

# client_id 为官网获取的AK， client_secret 为官网获取的SK
host = 'https://aip.baidubce.com/oauth/2.0/token?grant_type=client_credentials&client_id='输入ak'&client_secret='输入sk''
response = requests.get(host)
if response:
    print(response.json())
```

```python
# encoding:utf-8

import requests
import base64

'''
通用文字识别（高精度版）
'''

request_url = "https://aip.baidubce.com/rest/2.0/ocr/v1/accurate_basic"
# 二进制方式打开图片文件
f = open('5-1合同.png', 'rb')
img = base64.b64encode(f.read())

params = {"image":img}
access_token = '24.ca6450b3cd9ae038ebe13dbb5072e38c.2592000.1648104287.282335-25645440'
request_url = request_url + "?access_token=" + access_token
headers = {'content-type': 'application/x-www-form-urlencoded'}
response = requests.post(request_url, data=params, headers=headers)
if response:
    print (response.json())
```

```python
# encoding:utf-8

import requests
import base64

'''
智能财务票据识别
'''

request_url = "https://aip.baidubce.com/rest/2.0/ocr/v1/multiple_invoice"
# 二进制方式打开图片文件
f = open('5-2发票.png', 'rb')
img = base64.b64encode(f.read())

params = {"image":img}
access_token = '24.ca6450b3cd9ae038ebe13dbb5072e38c.2592000.1648104287.282335-25645440'
request_url = request_url + "?access_token=" + access_token
headers = {'content-type': 'application/x-www-form-urlencoded'}
response = requests.post(request_url, data=params, headers=headers)
if response:
    print (response.json())
```

```python
print(response.json()['words_result'][0]['result']['PurchaserName'][0]['word'])
print(response.json()['words_result'][0]['result']['TotalAmount'][0]['word'])
print(response.json()['words_result'][0]['result']['TotalTax'][0]['word'])
print(response.json()['words_result'][0]['result']['AmountInFiguers'][0]['word'])
```
