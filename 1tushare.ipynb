```python
###设置api接口,api接口用于与其他数据库对接，注册时会给予账号对应api信息，也决定所获得的权限，下面使用的是tushare网站给出的示例api接口，权限较低不过可以获取一些一般数据
###tushare网站过去完全免费，现在基本功能免费，部分高级功能对认证学生会员开放积分获取机制，可以尝试使用
pro = ts.pro_api('08e0747acd09a46936ae6ffb2a0d7fcaf4edaebad6aa33032481f5d0')
```

```python
###使用tushare网站示例程序拉取数据，简单拉取全部a股上市公司基本信息并存入pandas表df
df = pro.stock_basic(**{
    "ts_code": "",
    "name": "",
    "exchange": "",
    "market": "",
    "is_hs": "",
    "list_status": "",
    "limit": "",
    "offset": ""
}, fields=[
    "ts_code",
    "symbol",
    "name",
    "area",
    "industry",
    "market",
    "list_date",
    "fullname",
    "enname",
    "cnspell",
    "exchange",
    "curr_type",
    "list_status",
    "delist_date",
    "is_hs"
])
print(df)
```

```python
###行业排序,pandas库功能，对pandas表中列使用sort_values进行排序
df.sort_values('industry')
```
